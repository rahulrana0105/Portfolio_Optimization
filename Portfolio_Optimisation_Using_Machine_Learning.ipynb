```python
import numpy as np
import pandas as pd
```

```python
df.drop('HDFCLIFE.NS',axis=1,inplace=True)
df.drop('SBILIFE.NS',axis=1,inplace=True)

df.set_index('Date',inplace=True)

df.dropna(inplace=True)
df
```

```python
column_list = list(df.columns)
```

```python
def calculation_CAGR(col_list):
    
    start = df.iloc[0][col_list]
    end = df.iloc[-1][col_list]
    period = 5
        
    cagr = ((end/start)**(1/period)) - 1
    final_data['CAGR'] = cagr
    
    return final_data
```

```python
def volatility(col_list,data):
    temp_df = pd.DataFrame()
    temp_df[col_list] = data[col_list].pct_change()
    
    vol = temp_df[col_list].std(ddof=0) * np.sqrt(252)
    final_data['Volatility'] = vol
    
    return final_data
```

```python
def sharpe_ratio(col_list):
    sr = (final_data['CAGR'] - 0.06)/(final_data['Volatility'])
    final_data['Sharpe-ratio'] = sr
    return final_data
```

```python
final_data = pd.DataFrame(data = 0.0,index = column_list,columns =['CAGR','Volatility','Sharpe-ratio'])

final_data = calculation_CAGR(column_list)
final_data = volatility(column_list,df)

final_data = sharpe_ratio(column_list)
```

```python
df2 = pd.DataFrame()
for i in range(len(newlist)):
    df2[newlist[i]] = df[newlist[i]]
    
df2
```

```python
df3 = pd.DataFrame()
```

```python
# Calculating Log Returns

log_ret = np.log(df2/df2.shift(1))

df3 = log_ret.reset_index()
df3.drop('Date',axis=1,inplace=True)
df3.drop(0,inplace=True)
```

```python
df3
```

```python
# Initializing random weights

np.random.seed(42)

num_ports = 6000*100 
all_weights = np.zeros((num_ports, df3.shape[1]))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)
```

```python
# Obtaining the expected return, volatility & sharpe ratio of the portfolio corresponding to those weights

for x in range(num_ports):
    if x%6000 == 0:
        print(x)
    
    # Weights
    weights = np.array(np.random.random(15))
    weights = weights/np.sum(weights)
    
    # Save weights
    all_weights[x,:] = weights
    
    # Expected return
    ret_arr[x] = np.sum( (df3.mean() * weights * 252))
    
    # Expected volatility
    vol_arr[x] = np.sqrt(np.dot(weights.T, np.dot(df3.cov()*252, weights)))
    
    # Sharpe Ratio
    sharpe_arr[x] = ret_arr[x]/vol_arr[x]
```

```python
# MAXIMUM SHARPE RATIO FROM THE RANDOM WEIGHTS & THE WEIGHTS CORRESPONDING TO IT USING MAX & ARGMAX
print(sharpe_arr.max())
```

```python
#POINT p is the point of maximum sharpe ratio

p = sharpe_arr.argmax()
print(all_weights[p,:])
```

```python
mr = ret_arr[p]
mv = vol_arr[p]
```

```python
# Plotting the return & volatility of the portfolios on a Scatter Plot

import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.scatter(mv, mr,c='red', s=50) # red dot

plt.show()
```

```python
fy = np.linspace(0.2,0.35,50)
```

```python
def get_ret_vol_sr(weights):
    weights = np.array(weights)
    ret = np.sum(df3.mean() * weights) * 252
    vol = np.sqrt(np.dot(weights.T, np.dot(df3.cov()*252, weights)))
    sr = ret/vol
    return np.array([ret, vol, sr])

def min_vol(weights):
    return get_ret_vol_sr(weights)[1]

def check_sum(weights):
    #return 0 if sum of the weights is 1
    if (np.sum(weights) == 1):
        return 0
    
    return np.sum(weights)-1
```

```python
a = 1/15
c = ((0,1),)
b = []
d = ()
for i in range(15):
    b.append(a)
    d = d+c
```

```python
# Plotting the Efficient Frontier using SLSQP method 

from scipy.optimize import minimize
fx = []

for i in fy:
    print(i)
    cons = ({'type':'eq', 'fun':check_sum},
            {'type':'eq', 'fun': lambda w: get_ret_vol_sr(w)[0] - i})
    
    result = minimize(min_vol, b, method='SLSQP', bounds=d, constraints=cons)
    fx.append(result['fun'])
```

```python
plt.figure(figsize=(15,8))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.plot(fx,fy, 'r--', linewidth=3)
plt.show()
```
